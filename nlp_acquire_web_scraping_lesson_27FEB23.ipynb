## NY Times website

```python
respon = requests.get('https://www.nytimes.com/')
respon
```

```python
# what comes back ?
# sth messy big text box mash
# a lot of info can indicate that it's worthwhile

# R click 'inspect', or 'f12' key

soup = BeautifulSoup(respon.content, 'html.parser')
soup
```

```python
soup.find_all('h2')
```

```python
# finds all h3 elements on webpage : here, article titles...
# and some other stuff at the bottom ('<h3 class="css-1onhbft"', etc)

soup.find_all('h3')
```

```python
# focus on one of the classes

article_links = soup.find_all('h3', class_ = 'indicate-hover')
article_links
```

```python
# gets the title text for each article

[el.text for el in soup.find_all('h3', class_ = 'indicate-hover')]
```

```python
# # n33e the link to the h3 element

soup.find_all('a')

# everything that links to something (ie, '<a')
```

```python
all_a = soup.find_all('a', class_ = 'css-9mylee')
all_a
```

```python
all_a[0]['href']
# access just the 'href' piece : returns just the link
```

```python
# appending urls to a_list

a_list = []

for element in all_a:
    a_list.append(element['href'])
```
